### üßπ Data Cleaning: Handling `TotalCharges`

The `TotalCharges` column contains numeric values stored as strings, along with blank spaces representing missing or undefined entries.  
To ensure accurate analysis, these blank values are replaced with zero and the column is then converted to a numeric data type.

This step is essential for performing statistical analysis and visualizations involving total charges.


```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0")
df["TotalCharges"] = df["TotalCharges"].astype("float64")
```

### üîÑ Data Transformation: SeniorCitizen Mapping

The `SeniorCitizen` column is transformed by mapping its numeric values to more meaningful categorical labels.  
Values of `1` are converted to **"Yes"** and all other values are converted to **"No"**, improving readability and interpretability during analysis and visualization.


```python
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

# üìâ Churn Distribution Analysis

### Churn Distribution Using Pie Chart

```python
gb = df.groupby("Churn").agg(Count = ("Churn", "count"))
plt.pie(
    gb["Count"],
    labels = gb.index,
    autopct = "%1.1f%%",
    startangle = 90
)
plt.title("Percentage Of Churned Customer", fontsize = 14, fontweight = "bold")
plt.show()
```

### ‚úÖ Conclusion: Customer Churn Analysis

The analysis shows that approximately **26.5% of customers have churned**, while **73.5% of customers are retained**.  
This indicates that although a majority of customers continue using the service, a **significant portion is at risk of attrition**.

### üë• Gender-wise Churn Analysis

```python
ax = sns.countplot(
    x="gender",
    data=df,
    hue="Churn",
)
for container in ax.containers:
    ax.bar_label(container)

plt.title("Gender-wise Customer Churn Distribution", fontsize=14, fontweight="bold")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.legend(title="Churn Status")
plt.show()
```

### üìå Conclusion: Gender-wise Customer Churn

The gender-wise analysis shows that **customer churn is almost evenly distributed between male and female customers**.  
Both genders exhibit a similar number of churned and retained customers, indicating that **gender alone is not a significant driver of churn** in this dataset.

This suggests that customer attrition is likely influenced more by behavioral, contractual, or pricing factors rather than demographic factors such as gender.  
Future analysis should focus on variables like contract type, tenure, and monthly charges to better understand the key reasons behind customer churn.


### üë• Churn Distribution Across Senior and Non-Senior Customers


```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count Of Customer By Senior Citizen", fontsize=14, fontweight="bold")
plt.show()
```

```python
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize="index") * 100

ax = ct.plot(kind="bar", stacked=True, figsize=(8,6))

for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", label_type="center")

plt.title("Senior Citizen-wise Customer Churn (%)", fontsize=14, fontweight="bold")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage of Customers")
plt.legend(title="Churn Status")
plt.xticks(rotation=0)
plt.show()
```

### üìå Conclusion: Senior Citizen-wise Customer Churn

The analysis indicates a clear difference in churn behavior between senior and non-senior customers.  
Senior citizens exhibit a **significantly higher churn rate (41.7%)** compared to non-senior customers (**23.6%**).

This suggests that senior customers may face challenges related to service usability, pricing, or support, making them more likely to churn.  
Targeted engagement strategies, personalized plans, and improved customer support for senior citizens could help reduce churn and improve retention within this segment.



### ‚è≥ Customer Churn Across Tenure


```python
sns.histplot(
    data=df,
    x="tenure",
    hue="Churn",
    bins=72,
)

plt.title("Tenure Distribution by Churn Status", fontsize=14, fontweight="bold")
plt.xlabel("Tenure (Months)")
plt.ylabel("Customer Count")
plt.show()
```

### üìå Conclusion: Tenure-wise Customer Churn

The tenure distribution clearly shows that **customers with shorter tenure are more likely to churn**, with a high concentration of churned customers in the early months of service.  
As tenure increases, the number of churned customers decreases significantly, indicating stronger customer loyalty over time.

Long-term customers (higher tenure) are predominantly retained, suggesting that customer engagement and satisfaction improve with continued service usage.  
This insight highlights the importance of **early-stage retention strategies**, such as onboarding support, personalized offers, and proactive engagement, to reduce churn during the initial months.


### üìÑ Impact of Contract Type on Customer Churn


```python
ax = sns.countplot(
    x="Contract",
    hue="Churn",
    data=df,
    palette={"No": "#1f4e79", "Yes": "#a63603"}
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Customer Churn by Contract Type", fontsize=14, fontweight="bold")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.legend(title="Churn Status")
plt.xticks(rotation=15)
plt.show()
```

### üìå Conclusion: Contract-wise Customer Churn

The analysis shows that customers on **month-to-month contracts exhibit the highest churn**, indicating a greater likelihood of service discontinuation without long-term commitment.  
In contrast, customers with **one-year and two-year contracts demonstrate significantly lower churn rates**, with two-year contracts showing the strongest customer retention.

This pattern suggests that longer contract durations help build customer loyalty and reduce churn risk.  
Encouraging customers to move from month-to-month plans to longer-term contracts through incentives or bundled offers could be an effective strategy to improve overall customer retention.


### üßæ Column Overview

The `df.columns` attribute is used to display the list of all column names present in the dataset.  
This step helps in understanding the available features and serves as a reference for selecting variables during exploratory data analysis and modeling.


```python
df.columns
```

```python
cols = [
    "PhoneService", "MultipleLines", "InternetService",
    "OnlineSecurity", "OnlineBackup", "DeviceProtection",
    "TechSupport", "StreamingTV", "StreamingMovies"
]

fig, axes = plt.subplots(3, 3, figsize=(18, 14))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f"{col} Distribution", fontsize=12, fontweight="bold")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis="x", rotation=30)

plt.tight_layout()
plt.show()
```

### üìå Conclusion: Service-wise Customer Churn Analysis

The service-level analysis reveals strong relationships between customer churn and the type of services subscribed.  
Customers who **do not use value-added services** such as *Online Security, Online Backup, Device Protection,* and *Tech Support* exhibit **significantly higher churn rates** compared to those who subscribe to these services.

Among internet service types, customers using **Fiber Optic internet show higher churn** than those using DSL, indicating potential issues related to pricing or service expectations.  
Additionally, customers with **no internet service consistently show lower churn**, suggesting reduced dependency or fewer service-related concerns.

Streaming services (*Streaming TV* and *Streaming Movies*) show a relatively balanced churn pattern, implying that entertainment services alone are not strong churn drivers.  
Overall, the presence of **support and security-related services appears to play a crucial role in customer retention**.

These insights suggest that bundling security and support services, improving fiber-optic service quality, and promoting value-added offerings can be effective strategies to reduce customer churn.

The analysis shows that customers using **electronic check** as a payment method have the **highest churn rate**, indicating a higher likelihood of service discontinuation compared to other payment options.

This pattern implies that automatic payment methods may enhance customer retention by reducing payment friction and missed payments.  
Encouraging customers to switch from electronic checks to automated payment options could be an effective strategy to reduce overall churn.
